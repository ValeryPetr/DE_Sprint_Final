```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("yellow_taxi_data_cleaned.csv")
data1 = data[(data["trip_distance"] < 100) & (data["trip_distance"] >= 0)]
```

```python
data1.plot.scatter(x='trip_distance', y='tip_amount')
plt.savefig('tips_vs_distance.png')
plt.show()
```

```python
data1 = data[data["passenger_count"] > 0]
data1.plot.scatter(x='passenger_count', y='tip_amount').set_xticks(np.arange(0, 11, 1.0))
plt.savefig('tips_vs_passenger_count.png')
plt.show()
```
